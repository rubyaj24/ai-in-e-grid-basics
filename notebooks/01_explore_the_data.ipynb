# 01 — Explore the Data

Before we do any AI, let's get our hands dirty with the data.
Open this notebook, run each cell, and see what the numbers look like.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/smart_meter_data.csv', parse_dates=['timestamp'])
df.head()
```

### How many rows? How many days?

```python
print(f"Rows: {len(df)}")
print(f"Days: {len(df) / (24 * 4):.0f}")
print(f"Date range: {df['timestamp'].min()} to {df['timestamp'].max()}")
```

### Plot the whole dataset

See the daily pattern? The peaks and valleys repeat every day.

```python
plt.figure(figsize=(14, 4))
plt.plot(df['timestamp'], df['load_kw'], linewidth=0.5)
plt.title('60 days of smart meter data')
plt.ylabel('Load (kW)')
plt.tight_layout()
plt.show()
```

### Zoom in on one week

Now you can really see the morning and evening peaks.

```python
one_week = df.iloc[:7 * 24 * 4]
plt.figure(figsize=(12, 4))
plt.plot(one_week['timestamp'], one_week['load_kw'])
plt.title('One week of data — can you spot the weekend?')
plt.ylabel('Load (kW)')
plt.tight_layout()
plt.show()
```

### Temperature vs load

Let's see if there's a relationship between temperature and power usage.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['temperature_c'], df['load_kw'], alpha=0.1, s=1)
plt.xlabel('Temperature (°C)')
plt.ylabel('Load (kW)')
plt.title('Temperature vs Load')
plt.tight_layout()
plt.show()
```
