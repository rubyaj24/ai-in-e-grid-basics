**Goal**: Beat the benchmark models from the demos.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

df = pd.read_csv('../data/smart_meter_data.csv', parse_dates=['timestamp'])

# Standard feature engineering
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['dow_sin'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
df['dow_cos'] = np.cos(2 * np.pi * df['dayofweek'] / 7)

features = ['hour_sin', 'hour_cos', 'dow_sin', 'dow_cos', 'is_weekend', 'temperature_c']

split = int(len(df) * 0.8)
X_train = df[features].iloc[:split]
X_test = df[features].iloc[split:]
y_train = df['load_kw'].iloc[:split]
y_test = df['load_kw'].iloc[split:]

# Benchmark (Random Forest with default params)
benchmark = RandomForestRegressor(n_estimators=200, max_depth=8, random_state=42)
benchmark.fit(X_train, y_train)
benchmark_mae = mean_absolute_error(y_test, benchmark.predict(X_test))
print(f"Benchmark MAE: {benchmark_mae:.4f}")
```
